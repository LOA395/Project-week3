```python
import pandas as pd

# Lista de países a eliminar
paises_a_eliminar = ['Albania', 'Montenegro', 'Andorra', 'North Macedonia', 
                     'Bosnia and Herzegovina', 'Türkiye', 'Kosovo under UNSCR 1244/99', 
                     'Switzerland', 'Serbia', 'United Kingdom']

# Eliminar las filas donde el país en 'Country1' está en la lista de países a eliminar
df = df[~df['Country'].isin(paises_a_eliminar)]

# Verificar el DataFrame después de la eliminación
df
```

```python
import pandas as pd

# Reemplazar NaNs con el valor del año anterior
df_filled = df.fillna(method='ffill', axis=1)

# Verificar el resultado
df_filled
```

```python
import pandas as pd

# Convertir los datos del DataFrame
df_melted = df_filled.melt(id_vars=['Country'], var_name='Year', value_name='Average air pollution level')

# Ordenar el DataFrame
df_sorted = df_melted.sort_values(by=['Country', 'Year'])

# Mostrar el DataFrame resultante
print(df_sorted)
```

```python
import pandas as pd

# Asegurarse de que la columna 'Average air pollution level' es de tipo numérico
df_sorted['Average air pollution level'] = pd.to_numeric(df_sorted['Average air pollution level'], errors='coerce')

# Redondear los valores en la columna 'Average air pollution level' a dos decimales
df_sorted['Average air pollution level'] = df_sorted['Average air pollution level'].round(2)

# Verificar el resultado
print(df_sorted)
```

```python
import pandas as pd

# Ruta del archivo
ruta_xlsx = 'indice_desarollo_humano.xlsx'

# Leer el archivo Excel
datos = pd.read_excel(ruta_xlsx, engine='openpyxl')

# Mostrar las primeras filas del DataFrame
datos.head(10)
```

```python
import pandas as pd

# Eliminar y renombrar las columnas especificadas
datos = (
    datos.drop(columns=['Unnamed: 0', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6'])
         .rename(columns={'Unnamed: 1': 'Country', 'Unnamed: 2': 'IDH-2022'})
)

# Eliminar las primeras 6 filas del DataFrame
datos = datos.iloc[6:].reset_index(drop=True)

# Mostrar las primeras 10 filas del DataFrame resultante
datos.head(10)
```

```python
import pandas as pd

# Lista de países que necesitas
paises_necesarios = [
    'Austria', 'Belgium', 'Bulgaria', 'Cyprus', 'Czechia', 'Germany', 'Denmark', 
    'Estonia', 'Greece', 'Spain', 'Finland', 'France', 'Croatia', 'Hungary', 
    'Ireland', 'Iceland', 'Italy', 'Liechtenstein', 'Lithuania', 'Luxembourg', 
    'Latvia', 'Malta', 'Netherlands', 'Norway', 'Poland', 'Portugal', 
    'Romania', 'Sweden', 'Slovenia', 'Slovakia'
]

# Filtrar, ordenar y restablecer el índice
datos_ordenados = (
    datos[datos['Country'].isin(paises_necesarios)]   # Filtrar por países necesarios
    .sort_values(by='Country')                       # Ordenar alfabéticamente por 'Country'
    .reset_index(drop=True)                         # Restablecer el índice
)

# Mostrar el DataFrame resultante
datos_ordenados
```
